```python

# Match only grade-based metrics
keyword_mapping = {
    "ESG Combined Score": "ESG Combined Score Grade",
    "ESG Score": "ESG Score Grade",
    "Environmental Pillar Score": "Environmental Pillar Score Grade",
    "Social Pillar Score": "Social Pillar Score Grade",
    "Governance Pillar Score": "Governance Pillar Score Grade"
}

def extract_scores_by_keyword(filepath):
    score_by_year = {}
    try:
        for sheet in ["Environment", "Social", "Governance", "Controversies"]:
            df = pd.read_excel(filepath, sheet_name=sheet, header=None, engine="openpyxl")
            if df.shape[0] < 12:
                continue
            years = df.iloc[4, 2:7].dropna().astype(int)
            metric_names = df.iloc[11:, 1].reset_index(drop=True)
            values = df.iloc[11:, 2:7].reset_index(drop=True)

            for i, col in enumerate(values.columns):
                year = years.iloc[i]
                if year not in score_by_year:
                    score_by_year[year] = {}
                for metric, val in zip(metric_names, values[col]):
                    if pd.notna(metric):
                        metric_str = str(metric).strip()
                        for keyword in keyword_mapping:
                            if keyword in metric_str:
                                score_by_year[year][keyword] = val
    except Exception as e:
        print(f"Error reading {filepath}: {e}")
    return score_by_year

def correct_years(combined_data_path, esg_folder_path, output_path):
    df = pd.read_csv(combined_data_path)
    df["Year (1)"] = pd.NA
    matched_years = {}

    for idx, row in df.iterrows():
        identifier = row["Identifier"]
        filepath = os.path.join(esg_folder_path, f"ESG Table for {identifier}.xlsx")
        if not os.path.exists(filepath):
            continue

        year_scores = extract_scores_by_keyword(filepath)
        for year, scores in year_scores.items():
            match = True
            for keyword, csv_col in keyword_mapping.items():
                csv_val = str(row[csv_col]).strip()
                file_val = str(scores.get(keyword, "")).strip()
                if pd.isna(row[csv_col]) or csv_val != file_val:
                    match = False
                    break
            if match:
                df.at[idx, "Year (1)"] = year
                break

    # Ensure unique years per identifier
    for ident, group in df.groupby("Identifier"):
        seen = set()
        for idx in group.index:
            year = df.at[idx, "Year (1)"]
            if pd.isna(year):
                continue
            original_year = int(year)
            new_year = original_year
            step = 0
            while new_year in seen:
                step += 1
                new_year = original_year - step
            df.at[idx, "Year (1)"] = new_year
            seen.add(new_year)

    df.to_csv(output_path, index=False)
    print(f"Saved corrected file to: {output_path}")
```
